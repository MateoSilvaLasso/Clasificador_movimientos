```python

%pip install pandas
%pip install matplotlib
%pip install seaborn

import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configuración inicial para gráficos
sns.set(style="whitegrid")
```

```python


ruta_principal = "../data"


clases = ["sentarse", "pararse", "caminar hacia delante", "caminar hacia atras", "girar"]


datos_completos = []

for clase in clases:
    ruta_clase = os.path.join(ruta_principal, clase)
    for archivo in os.listdir(ruta_clase):
        if archivo.endswith(".csv"):
            ruta_archivo = os.path.join(ruta_clase, archivo)
            df = pd.read_csv(ruta_archivo)
            df["clase"] = clase 
            datos_completos.append(df)


df_datos = pd.concat(datos_completos, ignore_index=True)
df_datos.head()
```

```python


print("Resumen de los datos:")
print(df_datos.info())
print(df_datos.describe())
```

```python


plt.figure(figsize=(8, 6))
sns.countplot(x="clase", data=df_datos, order=clases, palette="viridis")
plt.title("Distribución de muestras por clase")
plt.xlabel("Clase")
plt.ylabel("Número de muestras")
plt.show()
```

```python


caracteristicas = ["x", "y", "z", "visibility"]

for caracteristica in caracteristicas:
    plt.figure(figsize=(8, 6))
    sns.kdeplot(data=df_datos, x=caracteristica, hue="clase", fill=True, common_norm=False, palette="muted")
    plt.title(f"Densidad de {caracteristica} por clase")
    plt.xlabel(caracteristica)
    plt.ylabel("Densidad")
    plt.show()
```

```python


df_caracteristicas = df_datos[["x", "y", "z"]]
correlacion = df_caracteristicas.corr()

plt.figure(figsize=(8, 6))
sns.heatmap(correlacion, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Matriz de correlación entre características")
plt.show()
```

```python


for caracteristica in caracteristicas:
    plt.figure(figsize=(8, 6))
    sns.boxplot(x="clase", y=caracteristica, data=df_datos, palette="pastel")
    plt.title(f"Outliers en {caracteristica} por clase")
    plt.xlabel("Clase")
    plt.ylabel(caracteristica)
    plt.show()


for caracteristica in caracteristicas:
    Q1 = df_datos[caracteristica].quantile(0.25)
    Q3 = df_datos[caracteristica].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    df_datos[caracteristica] = np.where(
        (df_datos[caracteristica] < limite_inferior) | (df_datos[caracteristica] > limite_superior),
        df_datos[caracteristica].median(),
        df_datos[caracteristica]
    )

print("Outliers manejados mediante reemplazo con la mediana.")
```
